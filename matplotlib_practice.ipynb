```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds=pd.read_csv(r"C:\Users\Mou\Desktop\finity\diamonds.csv")
diamonds
```

```python
sns.set_style("darkgrid", {'axes.grid' : True})
sns.lmplot(x= 'carat', y='depth', data=diamonds)
plt.show()
```

```python
sns.lmplot(x='carat', y='depth', data=diamonds, hue='color',palette='Set2')
plt.show()
```

```python
#strip plot

sns.stripplot(y= 'carat', data=diamonds,jitter=False)
plt.ylabel('color')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='carat', y='price', data=diamonds)
plt.ylabel('price')
plt.show()
```
